```python
from vae import VAE_3D
from torchsummary import summary

import torch
```

```python
params = {
          'batch_size': 1,
          'img_size': 128,
          'act_function': 'mish', 
          'dimension': 3, 
          'num_channels': 1,    
          'num_filters': 16, 
          'kernel_height': 3, 'kernel_width': 3, 
          'stride_c': 1, 'stride_p': 2,
          'kernel_c': 1, 'kernel_p': 2, 
          'output_classes': 5,
          }

params['kernel_depth'] = int(params['kernel_height'])

### 3 Dimension ###
HWD = params['img_size']
CH = params['num_channels']
BATCH = params['batch_size']

model_3D = VAE_3D(params)
summary(model_3D, torch.randn([BATCH, CH, HWD, HWD, HWD])) # 3D

### 2 Dimension ###
params['dimension'] = 2
model_2D = VAE_3D(params)
summary(model_2D, torch.randn([BATCH, CH, HWD, HWD])) # 2D
```
